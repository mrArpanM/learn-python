### **Division by a number with 9 at unit place**

⚒ **Divide 7 by 9**
$$
\frac{7}{9} ⇒ \frac{7}{9+1} = \frac{7}{10} =\frac{0.7}{1} ⇒ \frac{initial=7}{multiplier=1}  ⇒ 0.7̂
$$

Now start writing a number in right to left order until the sequence start repeating, beginning with the "initial" number & then write the next digit by multiplying it with the digit in heighest place , So, we'll write:
* As we have digit 7,so we'll write 7, carrying 0 ⇒ 7 | then.
* Will multiply the digit at higest place value with the multiplier, so 7*1 = 7, so we'll write 7, carrying 0 ⇒ 77
* As we can see the sequence is repeating itself, so we conclude that the answer is 0.7̂
